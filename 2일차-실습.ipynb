````python
#  

def AvgFilter(x):
    global k
    global preAvg    
    if preAvg != 0:
        k = k + 1
        alpha = (k - 1)/k
        avg = alpha*preAvg+(1-alpha)*x
        preAvg = avg        
        print(k,'[]:',avg)
    else:
        k = 1
        preAvg = x        
        print(k,'[]:',preAvg)
    
preAvg=0
AvgFilter(5)
AvgFilter(10)
AvgFilter(15)
AvgFilter(30)
````

````python
#   
def myCalc(num1, num2, op):
    if op=='+':
        return num1+num2
    elif op=='-':
        return num1-num2
    elif op=='*':
        return num1*num2
    elif op=='/':
        if num2==0:
            print('0  ')
            return 0
        else:
            return num1/num2
    else:
        print('  . ')
        return 0
print('===  ===')
op = input('(+,-,*,/)  :')
num1 = int(input(' :'))
num2 = int(input(' :'))
print(': ',myCalc(num1,num2,op))
````
