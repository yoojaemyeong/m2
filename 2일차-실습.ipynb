## **print 함수의 속성**

--- 

```python
#파이썬
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

> `*objects` : 출력할 객체들  
> `sep` : 출력할 여러 객체 사이에 삽입할 문자열(구분자) 정의 (예: `print("ok","no",sep=":")` )  
> `end` : 출력이 끝난 후 문자열(예: 줄바꿈 `print("hello",end="\n")` )  
> `file` : 출력 대상 설정(예: 파일, 소켓, stdout 등 `print("helo", file=sys.stdout)` )  
> `flush` : 출력을 강제로 보내기, True일 때 강제 플러시(기본은 False)  

---

### 문자열 formatting
* **{:형식지정}** 
> `{:>10}` : 오른쪽 정렬, 전체10칸으로 문자외에는 기본 빈공백  
> `{:<10}` : 왼쪽 정렬  
> `{:^10}` : 가운데 정렬  
> `{:.5f}` : 부동소수점으로 소수점아래 5자리 표시  
> `{:,}` : 천단위 쉼표 표시  

````python
print("정렬:{:>10}".format("right"))
print("정렬:{:<10}".format("left"))
print("정렬:{:^10}".format("center"))
print("소수점표시:{:.5f}".format(3.14156265359))
print("천단위: {:,}".format(10000000))
````

````python
# format형식
s1 = '2025'
s2 = '1'
s3 = '1'
print('{0}-{1:*>2}-{2:*<2}'.format(s1,s2,s3))
````

## 영문자는 ASCII 코드표에 의해 연산

### [ASCII 코드표](https://namu.wiki/w/%EC%95%84%EC%8A%A4%ED%82%A4%20%EC%BD%94%EB%93%9C)      
  

* ASCII 코드표에 의해 알파벳 문자를 숫자로 변환해주는 함수 : ord('문자')
  > ord('A')  
  > 결과: 65
* ASCII 코드표에 의해 숫자를 알파벳 문자로 변환해주는 함수 : chr(숫자)
  > chr(97)  
  > 결과: a  

````python
chr(97)
````

````python
ord('a')
````

````python
# ASCII :: 문자 -> 숫자
#대문자
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in range(0,len(alphabet)):
    if alphabet[i]=="Z":
        print("%c(%d)" % ( alphabet[i], ord(alphabet[i])))
    else:
        print("%c(%d)" % ( alphabet[i], ord(alphabet[i])), end=", ")
#소문자         
alphabet2 = alphabet.lower()        
for i in range(0,len(alphabet2)):
    if alphabet2[i]=="z":
        print("%c(%d)" % ( alphabet2[i], ord(alphabet2[i])))
    else:
        print("%c(%d)" % ( alphabet2[i], ord(alphabet2[i])), end=", ")       

#숫자를 알파벳 문자로
for i in range(97,123):
    if i==122:
        print("%d(%c)" % (i,chr(i)))
    else:
        print("%d(%c)" % (i,chr(i)), end=", ")
        
````

# 이동평균필터
* 이동평균필터의 배치식
    > 
* 이동평균필터의 재귀식
    > 
* 식의 전개
* 
* 

````python
#이동평균필터 함수 

def AvgFilter(x):
    global k
    global preAvg    
    if preAvg != 0:
        k = k + 1
        alpha = (k - 1)/k
        avg = alpha*preAvg+(1-alpha)*x
        preAvg = avg        
        print(k,'[평균필터]:',avg)
    else:
        k = 1
        preAvg = x        
        print(k,'[평균필터]:',preAvg)
    
preAvg=0
AvgFilter(5)
AvgFilter(10)
AvgFilter(15)
AvgFilter(30)
````

````python
# 팩토리얼 함수
def factorial(a):
    if a==1:
        return 1
    else:
        return factorial(a-1)*a

print('5! = ',factorial(5))
````

## 계산기 함수 구현
1. 계산을 위한 기호: +, -, *, /
2. 두 수를 입력
3. 결과

````python
# 계산기 함수 구현
def myCalc(num1, num2, op):
    if op=='+':
        return num1+num2
    elif op=='-':
        return num1-num2
    elif op=='*':
        return num1*num2
    elif op=='/':
        if num2==0:
            print('0으로 나눈 에러')
            return 0
        else:
            return num1/num2
    else:
        print('입력한 연산자는 없습니다. ')
        return 0
print('=== 계산기 ===')
op = input('입력방식(+,-,*,/) 중 선택:')
num1 = int(input('첫번째 숫자입력:'))
num2 = int(input('두번째 숫자입력:'))
print('결과: ',myCalc(num1,num2,op))
````

````python
# 비밀번호 규칙
# 1. 비밀번호 길이가 8글자 미만 False
# 2. 비밀번호 8글자 이상이면서 문자 또는 숫자로 구성
def checkPassword(pwd):
    if len(pwd) < 8:
        return False
    if pwd.isalnum():
        return True
    else:
        return False

#전역변수
password = input('새로운 비밀번호를 입력하세요')

if checkPassword(password):
    print('옳바른 비밀번호로 생성됨')
else:
    print('비밀번호 규칙에 맞지 않음')
````
